```python
# Menentukan distribusi normal/skewed pada kolom yang memiliki missing value

# Mengambil nama kolom yang memiliki missing value
missing_columns = data.columns[data.isnull().any()]

# Menyimpan hasil evaluasi distribusi di variabel hasil
hasil = []

# Melakukan pengecekan distribusi berdasarkan mean dan median
for column in missing_columns:
    series = data[column].dropna()
    mean = series.mean()
    median = series.median()
    selisih = abs(mean - median)

    if selisih <= 0.1 * abs(mean):
        # Jika selisih mean dan median <= 10% dari mean, maka distribusi normal
        status = "Normal"
    else:
        # Jika selisih mean dan median > 10% dari mean, maka distribusi skewed
        status = "Skewed"

    # Simpan nama kolom dan statusnya
    hasil.append((column, status))

# Membuat DataFrame dari hasil pengecekan
data_hasil = pd.DataFrame(hasil, columns=['Kolom', 'Status Distribusi'])

# Menampilkan isi dari data_hasil
data_hasil
```

**Observasi:**
* Jika ditinjau secara keseluruhan, nilai minimum dan maksimum tergolong masuk akal untuk setiap kolomnya.
* Untuk mengetahui apakah data pada suatu kolom terdistribusi normal atau skewed, maka digunakan pendekatan besaran mean dan median. Jika perbedaan nilai mean dan median berkisar di angka 10% dari nilai mean, maka data kolom tersebut terdistribusi normal. Jika perbedaannya di atas 10%, maka data kolom terdistribusi skewed.
* Oleh karena itu, berdasarkan hasil analisa menggunakan python, diperoleh bahwa ada 4 kolom yang terdistribusi normal, yaitu `Life expectancy`, `Total expenditure`, `Income composition of resources`, dan `Schooling`. Sisanya berjumlah 10 kolom termasuk ke dalam distribusi skewed (tidak normal).
* Berdasarkan analisa tersebut, maka dapat ditarik kesimpulan untuk handling missing value, yaitu:
  1. Untuk 4 kolom yang terdistribusi normal, maka handling missing value dilakukan dengan mengisi missing value dengan nilai mean.
  2. Untuk 10 kolom lainnya yang tidak terdistribusi normal (skewed), maka handling missing value dilakukan dengan mengisi missing value dengan nilai median.

# 4. Handling Missing Value

```python
# Membersihkan spasi di awal dan akhir nama kolom
data.columns = data.columns.str.strip()

# Mengganti spasi ganda di tengah menjadi spasi satu
data.columns = data.columns.str.replace(r'\s+', ' ', regex=True)

# Variabel symmetric_columns menampung nama kolom yang distribusi simetris (normal)
symmetric_columns = ['Life expectancy', 'Total expenditure', 'Income composition of resources', 'Schooling']

# Pengisian missing value dengan nilai mean untuk kolom dalam variabel symmetric_columns
for column in symmetric_columns:
    data[column] = data[column].fillna(data[column].mean())

# Variabel skewed_columns menampung nama kolom yang distribusi skewed (tidak normal)
skewed_columns = ['Adult Mortality', 'Alcohol', 'Hepatitis B', 'BMI', 'Polio', 'Diphtheria', 'GDP', 'Population', 'thinness 1-19 years', 'thinness 5-9 years']

# Pengisian missing value dengan nilai median untuk kolom dalam variabel skewed_columns
for column in skewed_columns:
    data[column] = data[column].fillna(data[column].median())
```
